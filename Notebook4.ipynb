```python
#Ejercicio 2

import flask
import json
from flask import request, jsonify
import mysql.connector

app = flask.Flask(__name__)

conn = mysql.connector.connect(user="root",host="localhost",password="realarsenal")
cursor = conn.cursor()
cursor.execute("USE valdivia")


@app.route('/', methods=['GET'])
def home():
    return '''<h1>API Valdivia</h1>
<p>Un prototipo de API para la base de datos Valdivia.</p>'''

@app.errorhandler(404)
def page_not_found(e):
    return "<h1>404</h1><p>No se pudo encontrar el recurso solicitado.</p>", 404

@app.route('/api_valdivia/v1/resources/personas', methods=['GET'])
def api_personas():
    ## Consultar MySQL para obtener datos sobre las personas
    result=cursor.execute('SELECT Nombre,Edad FROM persona;')
    all_films =cursor.fetchall()
    ## Conservar el nombre de los atributos
    row_headers=[x[0] for x in cursor.description]
    ## Transformar resultados en datos JSON
    json_data=[]
    for result in all_films:
        json_data.append(dict(zip(row_headers,result)))
    return jsonify(json_data)

@app.route('/api_valdivia/v1/resources/desplazarse', methods=['GET'])
def api_desplazarse():
    ## Consultar MySQL para obtener datos sobre los datos en la tabla desplazarse
    result=cursor.execute('SELECT pk_RUT,Latitud,Longitud FROM desplazarse;')
    all_films =cursor.fetchall()
    ## Conservar el nombre de los atributos
    row_headers=[x[0] for x in cursor.description]
    ## Transformar resultados en datos JSON
    json_data=[]
    for result in all_films:
        json_data.append(dict(zip(row_headers,result)))
    return jsonify(json_data)

app.run(debug=False,port=1234)
```

```python
#Ejercicio 3

Para poder desplegar mi API mas alla de mi computador seria necesario hacer uso de un servidor web remoto en 
el cual reciba o muestre el resultado de cada consulta, en este caso nuestro "host" ya no seria 
localhost y nuestro "user" tampoco seria root.

Tendremos que subir nuestro archivo en donde esta la API al servidor, podriamos crear un formulario 
basico de consulta de usuarios, entonces cada vez que completemos el formulario y le pidamos que realice
la consulta, este servidor hara la peticion http a la API y esta devolvera la respuesta en formato JSON
para luego mostrarla en la pagina.

Dado que todo esta hecho de forma remota, podremos hacer uso de nuestra API y realizar 
consultas desde cualquier computador.
```
