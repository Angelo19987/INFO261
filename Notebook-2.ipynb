```python
# ejercicio 4 n°5)
df = pd.read_sql('SELECT staff_id,avg(amount) as volumen FROM payment group by staff_id;', con=db_connection)
display(df)
```

```python
# ejercicio 4 n°6)
df = pd.read_sql('SELECT staff_id,sum(amount) as volumen FROM payment group by staff_id order by volumen desc limit 1 ;', con=db_connection)
display(df)
```

```python
#ejercicio 5 n°1)
df = pd.read_sql('SELECT title,name FROM film inner join language;', con=db_connection)
display(df)
```

```python
#ejercicio 5 n°2)
df = pd.read_sql('SELECT title,name FROM film inner join language LIMIT 20;', con=db_connection)
display(df)
```

```python
#ejercicio 5 n°3)
df = pd.read_sql('SELECT first_name,last_name,city FROM customer inner join city LIMIT 20 ;', con=db_connection)
display(df)
```

```python
#ejercicio 5 n°4)
df = pd.read_sql('SELECT title,count(title) as suma FROM film INNER JOIN inventory ON film.film_id = inventory.film_id INNER JOIN rental ON inventory.inventory_id = rental.inventory_id GROUP BY title LIMIT 10;', con=db_connection)
display(df)
```

```python
#ejercicio 5 n°5)
df = pd.read_sql('SELECT title,sum(amount) as suma FROM film INNER JOIN inventory ON film.film_id = inventory.film_id INNER JOIN rental ON inventory.inventory_id = rental.inventory_id INNER JOIN payment ON rental.rental_id = payment.rental_id GROUP BY title ORDER BY suma DESC LIMIT 1;', con=db_connection)
display(df)
```
